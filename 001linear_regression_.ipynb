```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 6, 141)
y = (x - 2.5) ** 2 - 1
plt.plot(x, y)
plt.show()
```

```python
def dF(theta):
    return 2 * (theta - 2.5)
```

```python
def F(theta):
    return (theta - 2.5) ** 2 - 1
```

```python
theta, epsilon, lr = 0.0, 1e-8, 0.1
```

```python
while True:
    grad = dF(theta)
    theta_ = theta
    theta -= lr * grad

    if abs(F(theta) - F(theta_)) < epsilon:
        break
theta_, F(theta_)
```

```python
theta, epsilon, lr = 0.0, 1e-8, 0.1
his_theta = []
while True:
    grad = dF(theta)
    theta_ = theta
    his_theta.append(theta)
    theta -= lr * grad

    if abs(F(theta) - F(theta_)) < epsilon:
        break
plt.plot(x, F(x))
plt.plot(np.array(his_theta), F(np.array(his_theta)), 'r', marker='+')
plt.show()
```

```python

def gradient_descent(init_theta, lr, epsilon=1e-8):
    theta = init_theta
    theta_his = []
    theta_his.append(init_theta)
    while True:
        grad = dF(theta)
        theta_ = theta
        theta -= lr * grad
        theta_his.append(theta)
        if abs(F(theta) - F(theta_)) < epsilon:
            break

def plot_theta_his():
    plt.plot(x, F(x))
    plt.plot(np.array(his_theta), F(np.array(his_theta)), 'r', marker='+')
    plt.show()
```

```python
lr = 0.01
theta_history = []
gradient_descent(0., lr)
plot_theta_his()
len(theta_history)
```
